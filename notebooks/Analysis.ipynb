```python
def remove_measurements(text):
    ## Takes in raw text of a line of ingredient. Removes amounts and returns a list of just the remaining noun phrases.

    measurements=['tsp','tbsp','knob','cup','fl oz','oz','lb']
    doc=nlp(text)
    doc=nlp(' '.join([re.sub(r'\d\w+','',word) for word in doc.text.split(' ') ]).strip()) #remove number and amount i.e. 80g
    doc=nlp(' '.join([re.sub(r'\([^)]*\)','',word) for word in doc.text.split(' ') ]).strip())
    doc=nlp(' '.join([word for word in doc.text.split(' ') if word not in [ent.text for ent in list(doc.ents) if ent.label_ =='CARDINAL']])) #remove numbers
    doc=nlp(' '.join([word for word in doc.text.split(' ') if word not in measurements]).strip()) #remove any remaining measurements
    nouns=[]
    for chunk in doc.noun_chunks:
        nouns.append(chunk.text)
    return(nouns)
```

```python
def transform_ingredients_list(ingredients_list):
    transformed_ingredients=[]
    for ingredient in ingredients_list:
        transformed_ingredients.append(remove_measurements(ingredient))
    return(transformed_ingredients)
```
